# Zahtjevi na funkciju procjene rizika pojedine konfiguracije

## Ulazni podaci i definicije

**A - atributi** - skup svih kolona iz baze podataka

**SZ - sigurnosni zahtjevi** - skup svih sigurnosnih zahtjeva, pri cemu je sigurnosni zahtjev skup atributa iz A koji govori koji atributi se ne bi smjeli pojavljivati zajedno kod jednog pruzatelja usluge

**PU - pruzatelji usluga** - skup svih dostupnih pruzatelja usluga

**S(SZ) - osjetljivost** - preslikavanje u kojem je svakom sigurnosnom zahtjevu dodijeljena osjetljivost (vaznost)

**Re(PU) - reputacija pruzatelja usluge** - preslikavanje u kojem je svakom pruzatelju usluge pridruzena reputacija (povjerenje u pruzatelja usluge)

**C(A) - konfiguracija** - smjestanje atributa u proizvoljan broj skupova (predstavlja grupiranje atributa)

**D(C) - distribucija** - pridruzivanje izmedu elemenata konfiguracije i pruzatelja usluga, pri cemu jedan pruzatelj usluga moze biti pridruzen s vise elemenata konfiguracije

** R(D(C), C(A), Re(PU), S(SZ), PU, SZ, A) - funkcija rizika **

## Zahtjevi na funkciju rizika

### Parcijalna uredenost

Izmedu rizika za dvije kofiguracije definiran je odnos (<=).

### Radi nad istim podacima

Rizici se mogu usporedivati ako i samo ako su izracunati nad istim podacima. To znaci da su rizici R1(D1(C1), C1(A1), Re1(PU1), S1(SZ1), PU1, SZ1, A1) i R2(D2(C2), C2(A2), Re2(PU2), S2(SZ2), PU2, SZ2, A2) usporedivi ako i samo ako vrijedi:

A1 = A2

SZ1 = SZ2

PU1 = PU2

S1(SZ1) = S2(SZ2)

Re1(PU1) = Re2(PU2)

### Fokus iskljucivo nad podacima

Sigurnost pojedine konfiguracije se mjeri samo u sigurnosti podataka. Razne izmjene programa koje se mogu dogoditi od strane napadaca ne racunaju se u sigurnost konfiguracije.

### Interval u kojem se moze nalaziti rizik

Rizik ne moze biti negativan!

Maksimalna vrijednost rizika mogu poprimiti bilo koju pozitivnu vrijednost. Tj. vrijednost 0 je minimalna vrijednost rizika - za promatranu konfiguraciju i distribuciju uz definirane atribute, sigurnosne zahtjeve, pruzatelje usluga, reputacije pruzatelja usluga i osjetljivost sigurnosnih zahtjeva prema kriterijima definiranim funkcijom rizika nema rizika.
Minimalnu vrijednost rizika nije uvijek moguce postici(zbog atributa, sigurnosnih zahtjeva, dostupnih posluzitelja...), pa za odredene slucajeve minimum postaje neka druga pozitivna vrijednost. 

Maksimalna vrijednost funkcije rizika nije definirana i nije nuzno da bude definirana. Maksimalna vrijednost rizika sigurno ce se postici ako kod najnepouzdanijeg pruzatelja usluge postavimo sve atribute, no limitiranje vrijednosti na neki interval znacilo bi da se funkcija mora nekako "spljostiti" na neki interval. Obicno se uzima da je taj interval [0, 1]. Kako bi se funkcija spljostila na neki interval mora se ili normalizirati tako da se svaka vrijednost rizika podijeli s maksimalnom, u tom slucaju dobivamo hrpu bespotrebnog dijeljena, ili bi se trebala koristiti neka fukcija koja bi bez poznavanja maksimalne vrijednosti skalirala rizik na taj interval. Ta funkcija moze biti linearna ili nelinearna, no u oba slucaja razlike u rizicima postaju manje ili se zbog nelinearnosti gube originalni odnosi zamisljeni formulom za rizik.
















## Potrebno definirati

### Izgled funkcije rizika uz konstantnu osjetljivost i promijenjivu reputaciju pruzatelja usluge

Funkcija rizika moze poprimiti 3 oblika koji imaju dobar trend. Potrebno se odluciti za jedan.


```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.pyplot import figure
import math
from collections import defaultdict

def show_risk_sensitivity_reputation_relationship_graph(rep_const=1.0):


   
    reputations = np.linspace(0.0, 1.0, 40, endpoint=True)

    risk_1 = []
    risk_2 = []
    risk_3 = []

    for rep in reputations:
        risk_1.append(-rep + 1)
        risk_2.append(0.5 ** rep)
        risk_3.append(math.log(-rep+2))

    fig, axs = plt.subplots(nrows=1, ncols=3, constrained_layout=True, figsize=(16,4))

    axs[0].plot(reputations, risk_1)
    axs[0].set_xlabel('Reputacija')
    axs[0].set_ylabel('Rizik')
    axs[0].set_title(f'Osjetljivost =const')

    axs[1].plot(reputations, risk_2 )
    axs[1].set_xlabel('Reputacija')
    axs[1].set_ylabel('Rizik')
    axs[1].set_title(f'Osjetljivost =const')


    axs[2].plot(reputations, risk_3)
    axs[2].set_xlabel('Reputacija')
    axs[2].set_ylabel('Rizik')
    axs[2].set_title(f'Osjetljivost =const')

show_risk_sensitivity_reputation_relationship_graph()
```

Istu stvar je potrebno definirati za slucaj kada je reputacija konstanta, a osjetljivost se mijenja.

Potrebno je definirati i odnos osjetljivosti i reputacije posluzitelj, tj. je li rizik manji kada prekrsimo zahtjev osjetljivosti 0.2 kod pruzatelja usluge reputacije 0.2, ili je manji kada prekrsimo zahtjev osjetljivosti 0.8 kod pruzatelja usluge reputacije 0.8.

### Primjer 1 
A = {x, y}

SZ = {{x}, {y}}

PU = {PU1, PU2}

S({x}) = S({y})

Re(PU1) < Re(PU2)

C1 = {{x, y}}; C2 = {{x}, {y}}

D1(C1) = [PU1: {{x, y}}, PU2: {0}]; D2(C1) = [PU1: {0}, PU2:{{x, y}}]

D3(C2) = [PU1: {{x}}, PU2: {{y}}]; D4(C2) = [PU1: {{y}}, PU2: {{x}}]

Uz sve navedeno, odnosi izmedu funkcije rizika moraju biti:

**R(D1) < R(D3) = R(D4) < R(D2)**

### Primjer 2

A = {x, y}

SZ = {{x}, {y}}

PU = {PU1, PU2}

S({x}) < S({y})

Re(PU1) < Re(PU2)

C1 = {{x, y}}; C2 = {{x}, {y}}

D1(C1) = [PU1: {{x, y}}, PU2: {0}]; D2(C1) = [PU1: {0}, PU2:{{x, y}}]

D3(C2) = [PU1: {{x}}, PU2: {{y}}]; D4(C2) = [PU1: {{y}}, PU2: {{x}}]

Uz sve navedeno, odnosi izmedu funkcije rizika moraju biti:

**R(D1) < R(D3) < R(D4) < R(D2)**

### Primjer 3

A = {x, y}

SZ = {{x,y}}

PU = {PU1, PU2}


Re(PU1) < Re(PU2)

C1 = {{x, y}}; C2 = {{x}, {y}}

D1(C1) = [PU1: {{x, y}}, PU2: {0}]; D2(C1) = [PU1: {0}, PU2:{{x, y}}]

D3(C2) = [PU1: {{x}}, PU2: {{y}}]; D4(C2) = [PU1: {{y}}, PU2: {{x}}]

Uz sve navedeno, odnosi izmedu funkcije rizika moraju biti:

** R(D3) = R(D4) < R(D1) < R(D2)**
