# Zahtjevi na funkciju procjene rizika pojedine konfiguracije

## Ulazni podaci i definicije

**A - atributi** - skup svih kolona iz baze podataka

**SZ - sigurnosni zahtjevi** - skup svih sigurnosnih zahtjeva, pri cemu je sigurnosni zahtjev skup atributa iz A koji govori koji atributi se ne bi smjeli pojavljivati zajedno kod jednog pruzatelja usluge

**PU - pruzatelji usluga** - skup svih dostupnih pruzatelja usluga

**S(SZ) - osjetljivost** - preslikavanje u kojem je svakom sigurnosnom zahtjevu dodijeljena osjetljivost (vaznost)

**Re(PU) - reputacija pruzatelja usluge** - preslikavanje u kojem je svakom pruzatelju usluge pridruzena reputacija (povjerenje u pruzatelja usluge)

**C(A) - konfiguracija** - smjestanje atributa u proizvoljan broj skupova (predstavlja grupiranje atributa)

**D(C) - distribucija** - pridruzivanje izmedu elemenata konfiguracije i pruzatelja usluga, pri cemu jedan pruzatelj usluga moze biti pridruzen s vise elemenata konfiguracije

** R(D(C), C(A), Re(PU), S(SZ), PU, SZ, A) - funkcija rizika **

## Zahtjevi na funkciju rizika

### Parcijalna uredenost

Izmedu rizika za dvije kofiguracije definiran je odnos (<=).

### Radi nad istim podacima

Rizici se mogu usporedivati ako i samo ako su izracunati nad istim podacima. To znaci da su rizici R1(D1(C1), C1(A1), Re1(PU1), S1(SZ1), PU1, SZ1, A1) i R2(D2(C2), C2(A2), Re2(PU2), S2(SZ2), PU2, SZ2, A2) usporedivi ako i samo ako vrijedi:

A1 = A2

SZ1 = SZ2

PU1 = PU2

S1(SZ1) = S2(SZ2)

Re1(PU1) = Re2(PU2)

### Fokus iskljucivo nad podacima###

Sigurnost pojedine konfiguracije se mjeri samo u sigurnosti podataka. Razne izmjene programa koje se mogu dogoditi od strane napadaca ne racunaju se u sigurnost konfiguracije.






### Primjer 1 
A = {x, y}

SZ = {{x}, {y}}

PU = {PU1, PU2}

S({x}) = S({y})

Re(PU1) < Re(PU2)

C1 = {{x, y}}; C2 = {{x}, {y}}

D1(C1) = [PU1: {{x, y}}, PU2: {0}]; D2(C1) = [PU1: {0}, PU2:{{x, y}}]

D3(C2) = [PU1: {{x}}, PU2: {{y}}]; D4(C2) = [PU1: {{y}}, PU2: {{x}}]

Uz sve navedeno, odnosi izmedu funkcije rizika moraju biti:

**R(D1) < R(D3) = R(D4) < R(D2)**



### Primjer 2

A = {x, y}

SZ = {{x}, {y}}

PU = {PU1, PU2}

S({x}) < S({y})

Re(PU1) < Re(PU2)

C1 = {{x, y}}; C2 = {{x}, {y}}

D1(C1) = [PU1: {{x, y}}, PU2: {0}]; D2(C1) = [PU1: {0}, PU2:{{x, y}}]

D3(C2) = [PU1: {{x}}, PU2: {{y}}]; D4(C2) = [PU1: {{y}}, PU2: {{x}}]

Uz sve navedeno, odnosi izmedu funkcije rizika moraju biti:

**R(D1) < R(D3) < R(D4) < R(D2)**


### Primjer 3

A = {x, y}

SZ = {{x,y}}

PU = {PU1, PU2}


Re(PU1) < Re(PU2)

C1 = {{x, y}}; C2 = {{x}, {y}}

D1(C1) = [PU1: {{x, y}}, PU2: {0}]; D2(C1) = [PU1: {0}, PU2:{{x, y}}]

D3(C2) = [PU1: {{x}}, PU2: {{y}}]; D4(C2) = [PU1: {{y}}, PU2: {{x}}]

Uz sve navedeno, odnosi izmedu funkcije rizika moraju biti:

** R(D3) = R(D4) < R(D1) < R(D2)**
